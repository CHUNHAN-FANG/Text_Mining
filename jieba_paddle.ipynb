## jieba 中文分词
[jieb官方github](https://github.com/fxsjy/jieba)

支持四种分词模式:<br>
* 精确模式
：将句子最精确地切开，适合文本分析
* 全模式
：将句子中所有可以成词的词语都扫描出来, 速度快，但不能解决歧义
* 搜索引擎模式
：在精确模式的基础上，对长词再次分割，提高召回率，适合搜索引擎分词
* paddle模式
：利用PaddlePaddle深度学习框架（飞桨_百度），训练序列标注（双向GRU）网络模型实现分词；支持词性标注。



```python
import numpy 
import numpy  as np
import codecs
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

#### jieba  四种模式示例

```python
# encoding=utf-8
import jieba

jieba.enable_paddle()# 启动paddle模式。 0.40版之后开始支持，早期版本不支持
strs=["我来到北京清华大学","乒乓球拍卖完了","中国科学技术大学","对于我来说学习python很快乐真的是"]
for str in strs:
    seg_list = jieba.cut(str,use_paddle=True) # 使用paddle模式
    print("Paddle Mode: " + '/'.join(list(seg_list)))

seg_list = jieba.cut("我来到北京清华大学", cut_all=True)
print("Full Mode: " + "/ ".join(seg_list))  # 全模式

seg_list = jieba.cut("我来到北京清华大学", cut_all=False)
print("Default Mode: " + "/ ".join(seg_list))  # 精确模式

seg_list = jieba.cut("他来到了网易杭研大厦")  # 默认是精确模式
print(", ".join(seg_list))

seg_list = jieba.cut_for_search("小明硕士毕业于中国科学院计算所，后在日本京都大学深造")  # 搜索引擎模式
print(", ".join(seg_list))
```

#### jieba_paddle模式_中文词性标注应用
利用paddle模式的词性标注，分词并挑选出在线旅游评论中的形容词，词频统计结果可侧面反映游客情感态度。


```python
# -*- coding: utf-8 -*-
name = 'ch_T-CY-UNIV'  # 已做“去除非中文词”处理的样本
path = 'data/迪士尼主题2021/'+name+'.txt'
file = codecs.open(path, 'r', 'utf-8')
content = file.read()
file.close()  ##释放内存
```

```python
print(content)
```

```python
import jieba
import jieba.posseg as pseg

jieba.enable_paddle() #启动paddle模式。0.40版之后开始支持，早期版本不支持

# # 框定完整特殊词汇
jieba.suggest_freq('data\words.txt', True)

#中文分词
segments = []
segs = pseg.cut(content, use_paddle = True ) ##当成长句整体  cut分词
for word, flag in segs:
    if flag == 'a' and len(word)>1:  # a形容词
        # print ( ' %s  %s '  % (word, flag))
        segments.append(word)
    else:
        pass
```

```python
segmentDF = pd.DataFrame({'segment':segments})

#根据停用词表 删除停用词
stopwords = [line.strip() for line in open('text/stopwords.txt', 'r', encoding='utf-8').readlines()]
segmentDF = segmentDF[~segmentDF['segment'].isin(stopwords)]
# 或手动输入停用词
segmentDF = segmentDF[~segmentDF['segment'].isin(['10','12','14','16','\r\n','\t'])]

segmentDF.head(60)
# #查看高词频
# segStat = segmentDF.groupby(by=["segment"] )["segment"].agg(["count"]).reset_index().sort_values(by="count", ascending=False)     ##agg=count（计数）方式   ##升序=False（降序）
# segStat.head(60)
```

```python
#查看高词频
segStat = segmentDF.groupby(by=["segment"] )["segment"].agg(["count"]).reset_index().sort_values(by="count", ascending=False)
segStat.to_excel('output/T-adj-CY-UNIV.xlsx')
segStat.head(60)
```
